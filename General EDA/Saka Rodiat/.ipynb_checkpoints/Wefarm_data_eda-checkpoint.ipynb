```python
# IMPORT THE NECESSARY LIBRARIES
import pandas as pd
import numpy as np
import duckdb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Path to the CSV file
csv_path = r"C:\Users\Lenovo\Desktop\DATA SCIENCE VOLUNTEER\farm_data.csv"
# Connect to DuckDB (in-memory)
# DuckDB allows querying large CSV/Parquet files efficiently without loading the full dataset into memory.
# The 'con' object will be used to execute queries and fetch results safely.
con = duckdb.connect()
print("Connection successful.")
```

```python
# Load first 5 rows safely 
df = con.execute(f"""
    SELECT *
    FROM read_csv_auto('{csv_path}')
    LIMIT 5
""").fetchdf()

print("First 5 rows:")
df
```

```python
total_rows = con.execute(f"""
    SELECT COUNT(*)
    FROM read_csv_auto('{csv_path}')
""").fetchdf()

print("Total rows:", total_rows.iloc[0,0])
```

```python
# This code converts a large CSV file into a Parquet file using DuckDB.
# DuckDB reads the CSV safely in chunks (without loading everything into memory),
# then writes it as a Parquet file, which is faster to load and more memory-efficient
# for subsequent data analysis
parquet_path = r"C:\Users\Lenovo\Desktop\DATA SCIENCE VOLUNTEER\farm_data.parquet"

con.execute(f"""
    COPY (SELECT * FROM read_csv_auto('{csv_path}')) 
    TO '{parquet_path}' (FORMAT PARQUET)
""")

print("CSV successfully converted to Parquet.")
```

```python
# Connect to DuckDB
con = duckdb.connect()
parquet_path = r"C:\Users\Lenovo\Desktop\DATA SCIENCE VOLUNTEER\farm_data.parquet"

selected_cols = [
    "question_language",
    "question_topic",
    "question_sent",
    "question_user_country_code",
    "question_user_created_at",
    "response_user_status",
    "response_user_created_at",
    "response_user_country_code"
]

# Random 1% sample
sample_df = con.execute(f"""
    SELECT {', '.join(selected_cols)}
    FROM parquet_scan('{parquet_path}')
    USING SAMPLE 1%
""").fetchdf()

print("Sample loaded successfully! Shape:", sample_df.shape)
```

```python
parquet_path = r"C:\Users\Lenovo\Desktop\DATA SCIENCE VOLUNTEER\farm_data.parquet"

columns_to_check = [
    "question_language",
    "response_language",
    "question_topic",
    "response_user_status",
    "question_user_country_code",
    "response_user_country_code"
    
]

for col in columns_to_check:
    unique_vals = con.execute(f"""
        SELECT DISTINCT {col}
        FROM parquet_scan('{parquet_path}')
        ORDER BY {col} ASC
    """).fetchdf()
    
    print(f"\nUnique values for {col} ({len(unique_vals)} total):")
    print(unique_vals)
```

**Which languages are predominantly used for questions?**

```python
# Visualization: Distribution of Question Language
# Count values
counts = sample_df['question_language'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(counts.index, counts.values)
plt.title("Distribution of Question Language")
plt.xlabel("Question Language")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

**The distribution shows that most questions are asked in English, making it the dominant language on the platform. This is followed by Swahili (swa), Nynorsk (nyn), and Luganda (lug).The presence of multiple regional languages indicates diverse user engagement across different language groups**

**WHICH STATUSES ARE MOST COMMON AMONG RESPONSE USERS?**

```python

# Visualization: Distribution of Response User Status
# Count values
counts = sample_df['response_user_status'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(counts.index, counts.values)
plt.title("Distribution of Response User Status")
plt.xlabel("Response User Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

**The Response User Status distribution shows that most responses come from live users, followed by zombie accounts, destroyed and blocked accounts. 
This gives insight into platform activity levels and the behavior of returning or inactive users.**

**Which countries do the users asking questions come from?**

```python


# Count values
counts = sample_df['question_user_country_code'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(counts.index, counts.values)
plt.title("Distribution of Question User Country Code")
plt.xlabel("Country Code")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

**The majority of users are from Kenya, followed by Uganda and Tanzania, while there is no representation from Great Britain (GB).**

**What are the 10 most common question topics among farmers?**



```python
top_topics = sample_df['question_topic'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=top_topics.values,
    y=top_topics.index,
    color="skyblue"  # Use 'color' instead of 'palette' for single-color bars
)
plt.title("Top 10 Question Topics")
plt.xlabel("Number of Questions")
plt.ylabel("Question Topic")
plt.show()
```

**The most frequently asked question topic is maize, followed by cattle and chicken, while goat and pig are the least common topics among the top 10.**
